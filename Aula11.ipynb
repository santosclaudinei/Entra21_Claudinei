```
'''Exercício 1 - (não usar o continue ou o break)
Crie um programa que mostre um memu com as seguites opções:

1) Soma
2) Subtração
3) Multiplicação
S) Para sair!

Para número 1: Peça 2 números e mostre a soma deles
Para número 2: Peça 2 númeors e mostre a subtração deles
Para númeor 3: Peça 2 números e mostre a multiplicação deles
Para S: Mostre uma mensagem de despedida e termine o programa.

Para qualquer outra opção deve aparecer a mensagem "Opção Inválida"'''
opcao = str('')
while opcao != 'S':
    print('''
############################################ SISTEMA ENTRA21 #################################################

    1 - realiza a SOMA (Vai solicitar dois numeros e para efetuar a soma e mostrar seu resultado)
    2 - realiza a SUBTRAÇÃO (Vai solicitar dois numeros e para efetuar a subtração e mostrar seu resultado)
    3 - realiza a Multiplicação (Vai solicitar dois numeros e para efetuar a multiplicação e mostrar seu resultado)
    S - FINALIZA o programa e mostra uma mensagem de despedida e termine o programa

###############################################################################################################''')
    opcao = str(input('Digite a opção desejada: '))
    if opcao == '1':
        num1 = input('Digite um numero: ')
        num2 = input('Digite outro numero: ')
        soma = int(num1) + int(num2)
        print('A soma entre {} e {} é {}' .format(num1, num2, soma))
    elif opcao == '2':
        num1 = input('Digite um numero: ')
        num2 = input('Digite outro numero: ')
        subtracao = int(num1) - int(num2)
        print('A subtração entre {} e {} é {}' .format(num1, num2, subtracao))
    elif opcao == '3':
        num1 = input('Digite um numero: ')
        num2 = input('Digite outro numero: ')
        multipicacao = int(num1) * int(num2)
        print('A multiplicação entre {} e {} é {}' .format(num1, num2,  multipicacao))
    elif opcao == 'S':
        print('Obrigado por utilizar o nosso sistema! Até a proxima')
    else:
        print('OPÇÃO INVALIDA!!!!')

```

```
''' Exercício 4 (use o conhecimento que foi passado no curso)
# Crie um programa com um menu interativo:

# Cadastro de pessoas v1.0
# A) Cadastrar Pessoa
# B) Ver todos os nomes cadastrados:
# C) Ver cadastro pelo nome:
# D) Apagar um cadastro pelo nome:
# S) Sair

# Para A: Cadastre os dados do cliente (Nome, idade, sexo, telefone
# Para B: Mostre todos os nomes dos clientes (só os nomes)
# Para C: Digite o nome do cliente e mostre todos os dados dele.
# Para D: Digite o nome do cliente e o apague do cadastro
# Para S: Mostre uma mensagem de despedida e termine o programa
# Para qualquer outro valor: Mostre "Opção invalida" '''


nome_cliente = []
idade_cliente = []
sexo_cliente = []
tel_cliente = []
escolha = ''
while (escolha != 'S'):
    print('''

############################################ Cadastro de pessoas v1.0 #################################################

A) Cadastrar Pessoa
B) Ver todos os nomes cadastrados:
C) Ver cadastro pelo nome:
D) Apagar um cadastro pelo nome:
S) Sair

#######################################################################################################################''')
    escolha = str(input('Escolha a opção desdejada dentre as listadas acima: ').upper().strip())
    if(escolha == 'A'):
        nome = str(input('Digite o seu nome completo: ').title().strip())
        nome_cliente.append(nome)
        idade = int(input('Digite a sua idade: '))
        idade_cliente.append(idade)
        sexo = str(input('Digite o seu sexo: ').upper())
        sexo_cliente.append(sexo)
        telefone = str(input('Digite seu numero de telefone: '))
        tel_cliente.append(telefone)
    elif(escolha == 'B'):
        for i in range (len(nome_cliente)):
            print(nome_cliente[i])
    elif(escolha == 'C'):
        nome_pesquisa = str(input('Digite o nome a pesquisar no banco de dados: ').title())
        for i in range (len(nome_cliente)):
            if (nome_pesquisa == nome_cliente[i]):
                print(f'''CLIENTE ENCONTRADO COM SUCESSO!!!

Nome: {nome_cliente[i]}
Idade: {idade_cliente[i]}
Sexo: {sexo_cliente[i]}
Telefone: {tel_cliente[i]} 
                ''')
                break
            elif i == (len(nome_cliente) - 1):
                print('Esse nome de cliente não está presente na nossa base de dados. ')
    elif(escolha == 'D'):
        nome_deletar = str(input('Digite o nome do cliente a ser removido do banco de dados: ').title())
        for j in range (len(nome_cliente)):
            if (nome_deletar == nome_cliente[j]):
                nome_cliente.remove(nome_cliente[j])
                idade_cliente.remove(idade_cliente[j])
                sexo_cliente.remove(sexo_cliente[j])
                tel_cliente.remove(tel_cliente[j])
                print('Cliente foi removido da nossa base de dados com sucesso!!!')
                break
            elif i == (len(nome_cliente)):
                print('Esse cliente não faz parte da nossa base de dados. ')

    elif(escolha == 'S'):
        print('Obrigado por usar o nosso sistema. Até a proxima.')
    else:
        print('Opção digitada é invalida!!!')
```
